```
import numpy as np
import os
```

```
# Plot
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```
def simple_plot():
    # Return random floats in the half-open interval [0.0, 1.0).
    data = np.random.random_sample((20,2))
    
    # plot only column 0
    generated_x_left = [i[0] for i in data]
    generated_y_left = [j[1] for j in data]

    plt.plot(generated_x_left, generated_y_left, "go", alpha=0.7)

    plt.xlabel('Velocity x left eye')
    plt.ylabel('Velocity y left eye')
    plt.xlim(-1.2,1.2)
    plt.ylim(-1.2,1.2)

    # bow
    plt.gca().set_aspect('equal', adjustable='box')

    plt.show()
    return data
    
data = simple_plot()
```

```
def two_plots(): 
    data = np.random.random_sample((20,4))
    fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(9,4))

    generated_x_left = [i[0] for i in data]
    # only get the y coordinate of left eye
    generated_y_left = [j[1] for j in data]
    ax1.plot(generated_x_left, generated_y_left, 'ro', alpha=0.7)

    generated_x_left = [i[2] for i in data]
    generated_y_left = [j[3] for j in data]    
    ax2.plot(generated_x_left, generated_y_left, 'go', alpha=0.7)


    ax1.set(xlabel='Velocity x left eye', ylabel='Velocity y left eye', 
            xlim=(-1.2,1.2), ylim=(-1.2,1.2), aspect='equal')
    ax2.set(xlabel='Velocity x right eye', ylabel='Velocity y right eye',
            xlim=(-1.2,1.2), ylim=(-1.2,1.2), aspect='equal')
    fig.suptitle("Title", ha="center", va="center")
    plt.show()

two_plots()
```
